```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values

# Select a linear model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction for Cyprus
X_new = [[37_655.2]]  # Cyprus' GDP per capita in 2020
X1=[[20_435]]
X2=[[46_657]]
Y1=model.predict(X1)
Y2=model.predict(X2)
M= (Y2[0][0]-Y1[0][0])/(X2[0][0]-X1[0][0])
C= Y1[0][0]-M*X1[0][0]
print(model.predict(X_new))

# Visualize the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 63_500, 4, 9])
#visualize the linear regression line
xpoints = np.array([20000, 68000])
y1=M*20000+C
y2=M*68000+C
ypoints = np.array([y1, y2])
plt.plot(xpoints, ypoints, color='red')

#taking input value for x
x_value= float(input("Enter a value: "))

#visualize the vertical line corresponding to value of x
x_vertical=np.array([x_value, x_value])
y=M*x_value+C
y_vertical=np.array([0,y])
plt.plot(x_vertical, y_vertical, color='green')

#visualize the horizontal line corresponding to value of y
x_horizontal=np.array([0, x_value])
y_horizontal=np.array([y, y])
plt.plot(x_horizontal, y_horizontal, color='green')


plt.show()

```
